### Data Visualization

When dealing with a set of data, often the first thing you’ll want to do is get a sense for how the variables are distributed.

The <b>distribution</b> of a statistical data set (or a population) is a listing or function showing all the possible values (or intervals) of the data and how often they occur. When a distribution of categorical data is organized, you see the number or percentage of individuals in each group. When a distribution of numerical data is organized, they’re often ordered from smallest to largest, broken into reasonably sized groups (if appropriate), and then put into graphs and charts to examine the shape, center, and amount of variability in the data.

The most convenient way to take a quick look at a univariate distribution in seaborn is the <b>distplot()</b> function. By default, this will draw a histogram and fit a kernel density estimate (KDE).

```python
live_numric_type = live.replace(
    {'status_type' : { 'video' : 1, 'photo' : 2, 'link' : 3, 'status' : 4 }})
live_numric_data = live_numric_type[
    ['status_type', 'num_reactions', 'num_comments', 'num_shares', 'num_likes',
     'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys']]

for ojha, feature in enumerate(list(live_numric_data.columns)[:-1]):
    fg = sns.FacetGrid(live_numric_data, hue='status_type', height=10)
    fg.map(sns.distplot, feature).add_legend()
    plt.show()
```

In descriptive statistics, a <b>box plot</b> or <b>boxplot</b> is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending from the boxes (whiskers) indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram. Outliers may be plotted as individual points. Box plots are non-parametric: they display variation in samples of a statistical population without making any assumptions of the underlying statistical distribution (though Tukey's boxplot assumes symmetry for the whiskers and normality for their length). The spacings between the different parts of the box indicate the degree of dispersion (spread) and skewness in the data, and show outliers. In addition to the points themselves, they allow one to visually estimate various L-estimators, notably the interquartile range, midhinge, range, mid-range, and trimean. Box plots can be drawn either horizontally or vertically. Box plots received their name from the box in the middle.

## Principal Component Analysis (PCA)

<b>Principal Component Analysis</b>, or <b>PCA</b> for short, is a method for reducing the dimensionality of data.
<br><br>
It can be thought of as a projection method where data with m-columns (features) is projected into a subspace with m or fewer columns, whilst retaining the essence of the original data.
<br><br>
The <b>PCA</b> method can be described and implemented using the tools of linear algebra.

```python
PCA_df = live
```

### Data Standardization 

<b>Data Standardization</b> is a data processing workflow that converts the structure of disparate datasets into a Common Data Format. As part of the Data Preparation field, Data Standardization deals with the transformation of datasets after the data is pulled from source systems and before it's loaded into target systems. Because of that, Data Standardization can also be thought of as the transformation rules engine in Data Exchange operations.
<br><br>
<b>Data Standardization</b> enables the data consumer to analyze and use data in a consistent manner. Typically, when data is created and stored in the source system, it's structured in a particular way that is often unknown to the data consumer. Moreover, datasets that might be semantically related may be stored and represented differently, thereby making it difficult for a data consumer to aggregate or compare the datasets.

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler
features = ['num_reactions', 'num_likes']

# Separating out the features
x=  PCA_df.loc[:,features].values

# Separating out the target
y = PCA_df.loc[: , ['status_type']].values

# Standardizing the features
x=StandardScaler().fit_transform(x)
print(x)
```

### PCA Projection to 2D

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

### Concatenating DataFrame along axis


```python
finalDf = pd.concat([principalDf, PCA_df[['status_type']]], axis = 1)
```

### Explained Variance Ratio

<b>Variance:</b> It is a measure of the variability or it simply measures how spread the data set is. Mathematically, it is the average squared deviation from the mean score. We use the following formula to compute variance var(x).


$$ var\left( x \right) = \frac{ \sum \left(x_i -\bar x\right)^2}{N}  $$

```python
print(pca.explained_variance_ratio_)
```

# Evaluation

<b>Clustering</b> is the task of dividing the population or data points into a number of groups such that data points in the same groups are more similar to other data points in the same group and dissimilar to the data points in other groups. It is basically a collection of objects on the basis of similarity and dissimilarity between them.

<b>Kmeans</b> algorithm is an iterative algorithm that tries to partition the dataset into Kpre-defined distinct non-overlapping subgroups (clusters) where each data point belongs to only one group. It tries to make the inter-cluster data points as similar as possible while also keeping the clusters as different (far) as possible. It assigns data points to a cluster such that the sum of the squared distance between the data points and the cluster’s centroid (arithmetic mean of all the data points that belong to that cluster) is at the minimum. The less variation we have within clusters, the more homogeneous (similar) the data points are within the same cluster.

$$ J = \sum_{i=1}^{m} \sum_{k=1}^{K} w_{ik} \Vert x^i - \mu_k \Vert^2$$

```python
from sklearn.cluster import KMeans
finalDf
```

```python
x = finalDf.iloc[:, [0,1]].values
```

### Elbow method

The <b>elbow method</b> is a heuristic method often used to truncate the number of parameters in data-driven models, e.g. number of principal components or clusters to describe dataset. It is a subjective method, and hence other approaches for determining the number of parameters may also used such as the silhouette method.

This method looks at the percentage of variance explained as a function of the number of clusters: One should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data. More precisely, if one plots the percentage of variance explained by the clusters against the number of clusters, the first clusters will add much information (explain a lot of variance), but at some point the marginal gain will drop, giving an angle in the graph. The number of clusters is chosen at this point, hence the "elbow criterion". 

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 

Error = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
ax.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of cluster')
plt.ylabel('Error')
plt.show()
```

<b>NOTE:</b>in the fogure we can see that the cluster number 4 is the optimum value of clusters.

# Accuracy Model

<b>Accuracy</b> is one metric for evaluating classification models. Informally, accuracy is the fraction of predictions our model got right.

$$ Accuracy = \frac{TP + TN}{TP + TN + FP + FN}  $$

<li><b> True Positive (TP) </b></li>
<li><b> False Positive (FP) </b></li>
<li><b> False Negative (FN) </b></li>
<li><b> True Negative (TN) </b></li>

### Confusion Matrix

A <b>confusion matrix</b> is a summary of prediction results on a classification problem. The number of correct and incorrect predictions are summarized with count values and broken down by each class. This is the key to the confusion matrix. The confusion matrix shows the ways in which your classification model is confused when it makes predictions.

```python
import graphviz
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

seed = 10

# Creating a LabelEncoder and fitting it to the dataset labels.
le = LabelEncoder()
le.fit(df['status_type'].values)

# Converting dataset str labels to int labels.
y = le.transform(df['status_type'].values)

# Extracting the instances data.
X = df.drop('status_type', axis=1).values

# Splitting into train and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.34, stratify=y, random_state=seed)
```

```python
tree = DecisionTreeClassifier(criterion='gini',
                              min_samples_leaf=5,
                              min_samples_split=5,
                              max_depth=None,
                              random_state=seed)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('DecisionTreeClassifier accuracy score: {}'.format(accuracy))
```

```python
print('Confusion Matrix is')
print(confusion_matrix(y_test, y_pred))
cm=confusion_matrix(y_test, y_pred)
plt.matshow(cm)
plt.show()
```

<b>Precision</b> (also called positive predictive value) is the fraction of relevant instances among the retrieved instances.
<br><br>
<b>Recall</b> (also known as sensitivity) is the fraction of the total amount of relevant instances that were actually retrieved.
<br><br>
Both precision and recall are therefore based on an understanding and measure of relevance.

```python
print(classification_report(y_test, y_pred) )
```
