### Implement K- Means Clustering Algorithm using iris Dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets
iris = datasets.load_iris()
headersName = ['sepal length', 'sepal width', 'petal length', 'petal width']
df = pd.read_csv(iris.filename)
```

```python
x = df.iloc[:, [0,1,2,3]].values
```

```python
# Exercise 1.1: Write the full description of KMeans algorithm and determine
# the minimum parameters which are necessary to execute it.
'''
- Kmeans algorithm is an iterative algorithm that tries to partition the
  dataset into Kpre-defined distinct non-overlapping subgroups (clusters)
  where each data point belongs to only one group. It tries to make the 
  inter-cluster data points as similar as possible while also keeping the 
  clusters as different (far) as possible. It assigns data points to a cluster 
  such that the sum of the squared distance between the data points and the 
  cluster’s centroid (arithmetic mean of all the data points that belong to 
  that cluster) is at the minimum. The less variation we have within clusters, 
  the more homogeneous (similar) the data points are within the same cluster.

- The minimum parameters which are necessary to execute it is 0 because all
  the parameters have a default value that will assigned to execute it.
'''
```

```python
'''
Now to start, let’s arbitrarily assign the value of k as 5.
We will implement kmeans clustering using k=5.
For this we will instantiate the KMeans class and assign it to
the variable kmeans5. 
'''
kmeans5 = KMeans(n_clusters = 5)
y_kmeans5 = kmeans5.fit_predict(x)
kmeans5.cluster_centers_
```

```python
'''
Repeat the experiment again and set the k as 2. Compare the outputs.
'''
kmeans2 = KMeans(n_clusters = 2)
y_kmeans2 = kmeans2.fit_predict(x)
kmeans2.cluster_centers_
```

```python
'''
There’s a method called the Elbow method, which is designed to help find
the optimal number of clusters in a dataset. So let’s use this method to 
calculate the optimum value of k. To implement the Elbow method, we need
to create some Python code (shown below), and we’ll plot a graph between
the number of clusters and the corresponding error value. This graph
generally ends up shaped like an elbow, hence its name: 
'''
Error = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of cluster')
plt.ylabel('Error')
plt.show()
```

```python
# Exercise 1.2: 
# From the plot above, can you determine the optimum number of clusters? 
'Yes'
# Why?
'because the no of cluster 3 is between the two change points (2 and 4)'
```

### Visualizing Clustering 

```python
'''
Let’s now visualize the clusters that were formed with the optimal k value.
You can clearly see these clusters in the image below, with each cluster
represented by a different color. Below is the scatter plot for 5 clusters
obtained early. 
'''
plt.scatter(x[:,0], x[:,1], c = y_kmeans5, cmap= 'rainbow') 
```

```python
# Exercise 1.3: plot the clusters with the optimum k? 
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(x)
plt.scatter(x[:,0], x[:,1], c = y_kmeans, cmap= 'rainbow') 
```
